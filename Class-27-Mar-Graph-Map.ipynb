```python
import folium as f
import pandas as pd
dfmap2=pd.read_csv('https://raw.githubusercontent.com/jtemporal/folium-101/main/cities.csv')
dfmap2
```

```python
len(dfmap2)
```

```python
my_map=f.Map(location=[13.133,16.10], zoom_start=2)
my_map
```

```python
city = dfmap2.loc[0]
city
```

```python
#add color to markers in map
def select_marker_color(row):
    if row['lived']=='yes':
        return 'pink'
    elif row['lived']=='no' and row['visited']=='yes':
        return 'purple'
    return 'blue'
```

```python
#apply select_marker_color to dfmap2
dfmap2['color']=dfmap2.apply(select_marker_color,axis=1)
```

```python
dfmap2.head()
```

```python
for _,city in dfmap2.iterrows():
    
    f.Marker(location=[city['latitude'],city['longitude']],
             popup=city['name'],
             tooltip=city['name'],
             icon=f.Icon(color=city['color'],prefix='fa',icon='circle')
            ).add_to(my_map)
    
my_map
```

```python
my_map.save('my_map.html')
```
